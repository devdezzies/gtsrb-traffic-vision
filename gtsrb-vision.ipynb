```python
import torch 
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt 
```

```python
# download classnames csv file - https://raw.githubusercontent.com/georgesung/traffic_sign_classification_german/master/signnames.csv
train_data = datasets.GTSRB(root="data", split="train", transform=transforms.Compose([
    # resize the image to 64x64
    transforms.Resize(size=(64, 64)),
    # transform to tensor
    transforms.ToTensor()
]), download=True)

test_data = datasets.GTSRB(root="data", split="test", transform=transforms.Compose([
    # resize the image to 64x64
    transforms.Resize(size=(64, 64)),
    # transform to tensor
    transforms.ToTensor()
]), download=True)
```

```python
train_dataloader = DataLoader(dataset=train_data, batch_size=32, shuffle=True)
a, b = next(iter(train_dataloader))
a, b
```
